```python
data_cleaned = data.drop(columns = ['Timestamp', 'Email address', 'What’s the most frustrating part of dining out for you?', 'What time of day do you usually prefer to eat out?', 'What city do you live in?'])

data_cleaned.head()
```

```python
yes_no_columns = ['Would you like to know live, real-time table availability and seating capacity at restaurants?', 'Would you be interested in pre-ordering your food 10-15 minutes before arriving at the restaurant given the menu is provided?', 'Would you find it helpful to know the waiting time or line-ups at the restaurant before you arrive?']

for col in yes_no_columns:
    data_cleaned[col] = data_cleaned[col].map({'Yes': 1, 'No': 0})
```

```python
data_cleaned.head()
```

```python
ordinal_mapping1 = {'Rarely': 0, 'Occasionally': 1, 'Frequently': 2, 'Always': 3}
ordinal_mapping2 = {'Yes': 1, 'No': 0, 'Depends': 2}


data_cleaned['How often do you dine out?'] = data_cleaned['How often do you dine out?'].map(ordinal_mapping1)
data_cleaned['How often do you go out without planning to dine out, but end up doing so?'] = data_cleaned['How often do you go out without planning to dine out, but end up doing so?'].map(ordinal_mapping1)

data_cleaned['Do you usually reserve a table before dining out?'] = data_cleaned['Do you usually reserve a table before dining out?'].map(ordinal_mapping2)
```

```python
data_cleaned.head()
```

```python
categorical_columns = ['What is your gender?', 'Which age group do you belong to?', 'On average, how long do you usually stay at a restaurant to dine?', 'If you reserve, how early do you usually book your table?']
data_encoded = pd.get_dummies(data_cleaned, columns=categorical_columns)

data_encoded.head()
```

```python
from ctgan import CTGAN
ctgan = CTGAN()

ctgan.fit(data_encoded, epochs = 300)

synthetic_data = ctgan.sample(10000)

synthetic_data.to_csv("synthetic_data10000res.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in data_encoded.columns:
    plt.figure(figsize=(10, 5))
    sns.kdeplot(data_encoded[col], label='Original Data', fill=True, alpha=0.5)
    sns.kdeplot(synthetic_data[col], label='Synthetic Data', fill=True, alpha=0.5)
    plt.title(f'Distribution of {col}')
    plt.legend()
    plt.show()
```

```python
from scipy.stats import ks_2samp

# Perform KS test for each column (comparing real vs. synthetic data)
for col in data_encoded.columns:
    real_data = data_encoded[col]
    synthetic_data_column = synthetic_data[col]
    ks_stat, p_value = ks_2samp(real_data, synthetic_data_column)
    
    print(f"KS Test for {col}: stat={ks_stat}, p-value={p_value}")
```
